```python
import sympy as sym

# declaring variables
x, y, z, a, b, c = sym.symbols('x, y, z, a, b, c')

f = sym.cos(x) / x
print(sym.limit(f, x, 90))

n = sym.symbols('n')
print(sym.limit(((n + 5) / (n + 3)) ** n, n, sym.oo))

# we take the derivative using diff
first = sym.diff(sym.sin(5 * x), x, 5) 
second = sym.diff(sym.cos(5*x), x, 5) 
third = sym.diff(sym.tan(5*x), x, 5)
fourth = sym.diff(sym.cot(5*x), x, 5)
print('1st derivative: ', first) 
print('2nd derivative: ', second) 
print('3rd derivative: ', third) 
print('4th derivative: ', fourth)

# notice the difference of the output format
sym.diff(sym.exp(-1) * sym.sin(5 + x))
```
