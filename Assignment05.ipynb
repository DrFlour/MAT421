```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as npl

x = np.linspace(-20,20,2000)
y = x * 20 + 8
plt.grid()
plt.plot(x,y)
plt.show()

def gradient_descent_1d(grad, cur_x=0.1, learning_rate=0.01, precision=0.0001, max_iters=10000):
    for i in range(max_iters):
        grad_cur = grad(cur_x)
        if abs(grad_cur) < precision:
            break  # When the gradient approaches 0, it is regarded as convergent
        cur_x = cur_x - grad_cur * learning_rate
        print("The", i, "time of iteration：x value is ", cur_x)
    print(" local minimum of x value=", cur_x)
    return cur_x

def grad_1d(x):
    return x 

gradient_descent_1d(grad_1d, cur_x=5, learning_rate=0.3, precision=0.0001, max_iters=1000)
```
