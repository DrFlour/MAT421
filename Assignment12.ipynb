```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X**5 + Y**5)
Z = X**5 + Y**5
ax.plot_surface(X, Y, Z, rstride=5, cstride=5, cmap='rainbow')
plt.show()

x = np.linspace(-5,5,500)
y = x ** 5 + 5
plt.grid()
plt.plot(x,y)
plt.show()
```
